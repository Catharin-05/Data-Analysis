# 1. Data Collection:

### Dataset Description

The dataset provided by an airline organization (referred to as "Invistico Airlines" for anonymity) contains feedback from customers who have previously flown with the airline. The dataset aims to provide insights into customer satisfaction and help predict whether future customers will be satisfied with the airline's services based on various parameters.

This dataset, sourced from Kaggle, consolidates customer feedback on multiple service aspects, along with their flight details. The primary goal is to identify key factors that influence customer satisfaction and provide the airline with actionable insights to enhance their services.

### Dataset Features:
The dataset consists of 129,880 records with the following features:

1. **satisfaction**: The target variable indicating whether a customer was satisfied or dissatisfied with their flight experience (`Satisfied` or `Dissatisfied`).
2. **Gender**: The gender of the customer (`Male` or `Female`).
3. **Customer Type**: Specifies whether the customer is a `Loyal` or `Disloyal` flyer.
4. **Age**: The age of the customer (numerical value).
5. **Type of Travel**: Indicates the purpose of the travel (`Business travel` or `Personal travel`).
6. **Class**: The travel class of the customer (`Business`, `Economy`, or `Economy Plus`).
7. **Flight Distance**: The distance flown by the customer during the flight (numerical value in miles).
8. **Seat comfort**: Rating for seat comfort (on a scale of 1-5).
9. **Departure/Arrival time convenient**: Rating for the convenience of departure and arrival times (on a scale of 1-5).
10. **Food and drink**: Rating for the quality of food and drink served (on a scale of 1-5).
11. **Gate location**: Rating for the convenience of the gate location (on a scale of 1-5).
12. **Inflight wifi service**: Rating for the inflight Wi-Fi service (on a scale of 1-5).
13. **Inflight entertainment**: Rating for the inflight entertainment options (on a scale of 1-5).
14. **Online support**: Rating for the airline's online support (on a scale of 1-5).
15. **Ease of Online booking**: Rating for the ease of booking online (on a scale of 1-5).
16. **On-board service**: Rating for the onboard service (on a scale of 1-5).
17. **Leg room service**: Rating for the leg room provided (on a scale of 1-5).
18. **Baggage handling**: Rating for the baggage handling process (on a scale of 1-5).
19. **Checkin service**: Rating for the check-in service (on a scale of 1-5).
20. **Cleanliness**: Rating for the cleanliness of the flight (on a scale of 1-5).
21. **Online boarding**: Rating for the online boarding experience (on a scale of 1-5).
22. **Departure Delay in Minutes**: The delay in departure time (numerical value in minutes).
23. **Arrival Delay in Minutes**: The delay in arrival time (numerical value in minutes). This field contains some missing values.

## 2.2 Outliers

1. Calculate Q1 and Q3:

- Q1 is the value below which 25% of the data falls, while Q3 is the value below which 75% of the data falls.
Determine the IQR:

2. IQR is calculated as:
- $ R=Q3‚àíQ1 $
3. Identify Outliers:

- Outliers are defined as values that fall below $  ùëÑ1‚àí1.5*IQR $ or $  ùëÑ3+1.5*IQR $

```python
numerical_cols = Invistico_Airline.select_dtypes(include=['int64', 'float64']).columns

# Plot boxplots for each numerical column
plt.figure(figsize=(12, 8))
for i, col in enumerate(numerical_cols):
    plt.subplot((len(numerical_cols) + 2) // 3, 3, i + 1)  # Arrange plots in rows and columns
    sns.boxplot(Invistico_Airline[col])
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()

plt.show()
```

This absence of outliers indicates that the numerical features maintain a consistent range of values without extreme variations, which can often skew data analysis and impact subsequent modeling efforts. The stability of the dataset enhances the reliability of insights derived from it, allowing us to focus on exploring the relationships between different features and their effects on sales and customer churn without the complicating influence of outliers.

## 2.3 Feature Engineering


```python
# Customer Lifetime Value CLV=(AveragePurchaseValue)√ó(PurchaseFrequency)√ó(CustomerLifespan)
# Customer Engagement Metrics based on number of emails opened, app usage frequency, and website visit durations.
# Product Affinity Score: Measure the likelihood of customers purchasing specific products based on their transaction history.
# Customer Segment
```

#### 2.3.1 Total Delay:

Description: This feature aggregates both Departure Delay in Minutes and Arrival Delay in Minutes to capture the total delay experienced during the flight. It is calculated as

$ Total Delay = Departure Delay in Minutes + Arrival Delay in Minutes $

```python
Invistico_Airline['Total Delay'] = Invistico_Airline['Departure Delay in Minutes'] + Invistico_Airline['Arrival Delay in Minutes']
```

By combining both types of delays, we get a comprehensive measure of the overall flight delay experienced by the customer. Analyzing how this total delay impacts customer satisfaction will help the airline identify the critical delay thresholds that influence negative reviews, enabling them to prioritize operational efficiency to minimize delays.


#### 2.3.2 Delay Indicator:

Description: This feature indicates whether a customer experienced any significant delay during their flight. It is a binary variable, where 1 represents a delayed flight (based on a threshold, such as >15 minutes) and 0 indicates no significant delay.

```python
Invistico_Airline['Experienced Delay'] = (Invistico_Airline['Total Delay'] > 15).astype(int)
```

The Delay Indicator provides a straightforward way to segment customers based on delay experiences. This can be cross-referenced with the satisfaction column to explore how the presence of a delay influences customer satisfaction and pinpoint how sensitive customers are to flight delays.
 
#### 2.3.3 Loyal Customer:
 
Description: This feature simplifies the Customer Type variable by converting it into a binary indicator where 1 denotes loyal customers and 0 denotes disloyal customers.

```python
Invistico_Airline['Loyal Customer'] = Invistico_Airline['Customer Type'].apply(lambda x: 1 if x == 'Loyal Customer' else 0)
```

Loyalty is often a key driver of satisfaction and business value. By analyzing satisfaction levels separately for loyal and disloyal customers, the airline can understand if loyalty programs are yielding positive experiences and which services help retain frequent flyers.


#### 2.3.4 Age group binning:

his feature groups customers into different age categories (e.g., 18-30, 31-45, 46-60, 61+) using the Age feature. The age categories can be labeled as "Young Adults," "Middle-aged," "Seniors," etc.

```python
Invistico_Airline['Age Group'] = pd.cut(Invistico_Airline['Age'], bins=[0, 20, 40, 60, 80, 100], labels=['0-20', '21-40', '41-60', '61-80', '81-100'])
```

Binning the age groups allows for the identification of patterns and trends in customer satisfaction across different age demographics. This analysis will help the airline tailor its services and marketing strategies to better meet the expectations of various age groups, such as providing different inflight entertainment or catering options based on age.

#### 2.3.5 Class Satisfaction:

Description: This feature combines the Class and satisfaction features to create a new categorical variable that captures satisfaction levels for different travel classes (e.g., "Business-Satisfied", "Economy-Unsatisfied").

```python
Invistico_Airline['Class_Satisfaction'] = Invistico_Airline['Class'] + '_' + Invistico_Airline['satisfaction']
```

This feature will allow the airline to assess whether satisfaction levels differ significantly across travel classes. The analysis can reveal if higher-paying customers (e.g., Business class) are more satisfied than Economy class passengers, and which class-specific services require improvement.

#### 2.3.6 Service Quality Score:

Description: This feature is a composite score created by averaging or summing various service-related ratings such as Seat comfort, Food and drink, Inflight entertainment, Leg room service, etc.

```python
service_cols = ['Seat comfort', 'Inflight entertainment', 'Food and drink', 'On-board service', 'Leg room service', 'Checkin service']
Invistico_Airline['Service Quality Score'] = Invistico_Airline[service_cols].mean(axis=1)
```

A single score representing the overall service quality simplifies the analysis of multiple service aspects. Correlating the Service Quality Score with satisfaction can reveal which customers are more satisfied overall and help identify which services have the most significant impact on a positive experience.

#### 2.3.7 Flight Distance Binning:

Description: This feature groups the Flight Distance variable into categories like "Short-haul," "Medium-haul," and "Long-haul" based on distance thresholds.

```python
Invistico_Airline['Flight Distance Category'] = pd.cut(Invistico_Airline['Flight Distance'], bins=[0, 1000, 3000, 10000], labels=['Short-haul', 'Medium-haul', 'Long-haul'])
```

Different flight lengths can have varying impacts on customer satisfaction. Long-haul flights may have stricter demands for comfort and service compared to short-haul flights. By analyzing satisfaction within these distance bins, the airline can optimize services based on flight length to meet customer expectations more effectively.

#### 2.3.8 Booking Convenience Score

Description: This feature aggregates the Ease of Online booking and Online boarding features into a single score to represent the convenience of the booking and boarding process.

```python
Invistico_Airline['Booking Convenience Score'] = Invistico_Airline[['Ease of Online booking', 'Online support', 'Online boarding']].mean(axis=1)
```

The booking and boarding experience is an essential part of customer satisfaction. A high convenience score can signal seamless user interactions, while a low score may indicate issues in the digital customer journey. Analyzing this score in relation to satisfaction will help the airline identify areas for improvement in their booking systems.


#### 2.3.9 Satisfaction Binarization

This feature converts the satisfaction variable into a binary outcome where 1 represents "Satisfied" customers and 0 represents "Dissatisfied" customers.

```python
Invistico_Airline['satisfaction'] = Invistico_Airline['satisfaction'].apply(lambda x: 1 if x == 'satisfied' else 0)
```

Binarizing satisfaction simplifies predictive modeling tasks by turning customer satisfaction into a classification problem. This transformation is useful for machine learning models that aim to predict whether a customer will be satisfied based on the input features. It also allows for clearer insight into the percentage of satisfied versus dissatisfied customers.
